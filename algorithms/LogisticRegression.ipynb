```python
import numpy  as np
import matplotlib.pyplot as plt 
```

```python
x = np.linspace(-10,10,200)

def _sigmoid(x):
    return 1/(1+np.exp(-x))

y = [_sigmoid(xi) for xi in x]
```

```python
plt.figure(figsize=(16,8))
plt.plot(x,y, 'go')
plt.axhline(y=1, color='blue', linewidth=2)
plt.axhline(y=0, color='red', linewidth=2)
plt.axvline(x=0)
plt.title('Sigmoid function')
plt.text(-10,0.05,'y = 0')
plt.text(-10,0.95,'y = 1')
plt.text(0.2,0.3,'x=0')

```
